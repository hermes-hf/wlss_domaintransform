```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../outputs.csv")
df = df[['sigma_s','sigma_r','alpha_r','psnr','ssim']].groupby(['sigma_s','sigma_r','alpha_r']).min()
df = df.reset_index()
df_psnr = df[['sigma_s','sigma_r','alpha_r','psnr']]
df_ssim = df[['sigma_s','sigma_r','alpha_r','ssim']]
psnr_series = pd.pivot(df_psnr, columns = ['sigma_r'], values =['psnr'], index = ['sigma_s','alpha_r']).reset_index()
ssim_series = pd.pivot(df_ssim, columns = ['sigma_r'], values =['ssim'], index = ['sigma_s','alpha_r']).reset_index()
```

# PSNR PLOTS

```python
psnr_series50 = psnr_series[psnr_series['alpha_r']==0.5].copy()
psnr_series75 = psnr_series[psnr_series['alpha_r']==0.75].copy()
psnr_series100 = psnr_series[psnr_series['alpha_r']==1.0].copy()
```

```python
series = psnr_series50.copy()
for i in range(len(series)):
    X = list(series.iloc[i]['psnr'].index.get_level_values('sigma_r'))
    Y = list(series.iloc[i]['psnr'])
    plt.plot(X,Y)
```

# SSIM PLOTS

```python
ssim_series50 = ssim_series[ssim_series['alpha_r']==0.5].copy()
ssim_series75 = ssim_series[ssim_series['alpha_r']==0.75].copy()
ssim_series100 = ssim_series[ssim_series['alpha_r']==1.0].copy()
```

```python
series = ssim_series50.copy()
for i in range(len(series)):
    X = list(series.iloc[i]['ssim'].index.get_level_values('sigma_r'))
    Y = list(series.iloc[i]['ssim'])
    plt.plot(X,Y)
```
